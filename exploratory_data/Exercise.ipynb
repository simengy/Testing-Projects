## Question 1:

Given the below subset of Uber’s schema, write executable SQL queries to answer the two questions below. Please answer in a single query and assume read only access to the database (i.e. do not use CREATE TABLE).

1. Between Oct 1, 2013 at 10am PDT and Oct 22, 2013 at 5pm PDT, what percentage of requests made by unbanned clients each day were canceled in each city?
2. For city_ids 1, 6, and 12, list the top three drivers by number of completed trips for each week between June 3, 2013 and June 24, 2013.

**Answer 1.1:**

```python
WITH t1 as
(
    SELECT X.city_id, X.date, COUNT(X.id) as n
    FROM
    ( 
        SELECT *, date(request_at::DATE AT TIME ZONE 'PDT') as date 
        FROM trips
        WHERE status = 'cancelled_by_client'
        AND request_at >= '2013-10-01 10:00:00'::DATE AT TIME ZONE 'PDT'
        AND request_at < '2013-10-22 17:00:00'::DATE AT TIME ZONE 'PDT'
    ) X
    INNER JOIN
    (
        SELECT * FROM users 
        WHERE banned = FALSE
        and role = 'client'
    ) Y
    ON (X.client_id = Y.user_id)
    GROUP BY 
    X.city_id, X.date
)
SELECT city_id, date, 100.0 * (0.0+n) / (SELECT COUNT(*) FROM trips) as percentage
FROM t1;

```

**Answer 1.2:**

```python
WITH t1 AS
(
    SELECT X.driver_id, week, COUNT(X.id) as count
    FROM
    (
        SELECT *, 
        date_trunc('week', TO_TIMESTAMP(actual_eta)::DATE AT TIME ZONE 'PDT') as week
        FROM trips
        WHERE status = 'completed'
        AND TO_TIMESTAMP(actual_eta) >= '2013-06-03'::DATE AT TIME ZONE 'PDT'
        AND TO_TIMESTAMP(actual_eta) < '2013-06-24'::DATE AT TIME ZONE 'PDT'
        AND city_id IN (1,6,12)
    ) X
    INNER JOIN
    (
        SELECT * FROM users
        WHERE role = 'driver'
    ) Y
    ON (X.driver_id = Y.user_id)
    GROUP BY driver_id, week
)
SELECT driver_id, week, r
FROM
(
    SELECT driver_id, week, 
    row_numbers() OVER (partition by week order by count DESC) as r
    FROM t1
) A
WHERE r<=3;
```

**Answer 2.3:**

We can use **A/B tesing** on KPI **"Average number of successful invites per user"** to test whether the new feature works or not:

Split the current users into placebo (Free Ride) and treatment (Surge Protector) groups in which the placebo users will receive a free ride for every suceesuful invite while the treatment will get a surge protector. Then let's run this campaign for a period of time, such as two weeks, and collect the data in the following format:

* INVITE_ID, INVITE_SENT_TIMESTAMP, SUCCUESS, INCENTIVE
* 1,  2016-01-02, 0, Surge
* 2,  2016-01-03, 1, Free
* 3,  2016-01-05, 0, Free
* ......
* 100, 2016-02-04, 0, Surge
* ......

In order to check the effectivness of new feautre, we can use right-tailed T-test to test the **null hypothesis**: the surge Protector group does not exhibit a higher **''average number of sucessful intives per user''** than Free-ride group. By comparing the computed p-value with different condfidence levels (1%, 5% or 10%), the product manager can make a decison of the new feature in production based on his/her belief (different p-value thresholds). For example, if p-values = 0.4 > chosen threshold = 5%, we shall accept the null hypothsis which means the new feature does not work, and vice versa.

**Answer 3.2: Building a predictive model**

The model built here is to predict the retained probability 6 months later (2014/01 - 2014-06). For other timeframe, we might need retrain the model.

First thing is to take care of missing values in 'avg_rating_by_driver', 'avg_rating_of_driver' and 'phone'. We imputed the missing values following, instead of removing:

* 'avg_rating_by_driver': minimum rating
* 'avg_rating_of_driver': minimum rating
* 'phone': 'other'

Moreover the categorical variables 'last_trip_date', 'city', 'phone' and 'singup_date' are dummy coded:

* "last_trip_date" is used to build the target 'Active' for the predictive model.
* 3 levels of 'city' are dummy coded as binary features. 
* There are ~400 'phone' are missing which has been imputed as 'other'.
* per "signup_date", we calculated the time between "data_pull_data" (2014-07-01) and "signup_date" to indicate the history of users started to use Uber service.

Furthermore, **Logistic Regression** and **Random Forest** are the two classifiers chosen. We have employed the variable importance from **Random Forest** to rank the most predictive features. Instead of Cross validation, we applied **Train-Test Split** to validate the model performances for this relatively large data (50K records). Specifically speaking, the classifiers are trained on 70% of data and then test on the rest of 30%. 

Last but not the least, The key performance metric is **AUC**. We can see **Random Forest (AUC = 0.85)** out-performs **Logistic Regression (AUC = 0.76)**. As for cross validation, we are expecting to have to similar results.

* Missing values filling and feature engineering

```python
data['avg_rating_by_driver'] = data['avg_rating_by_driver'].fillna(data['avg_rating_by_driver'].min())
data['avg_rating_of_driver'] = data['avg_rating_of_driver'].fillna(data['avg_rating_of_driver'].min())
data['phone'] = data['phone'].fillna('other')

data['user_history'] = pd.to_datetime(data['signup_date']).\
apply(lambda x: (pull_date - x).days)
```

* Dummy coding of city and phone

```python
data = data.join(pd.get_dummies(data['city']))
data = data.join(pd.get_dummies(data['phone']))
```

```python
features = [u'avg_dist',u'avg_rating_by_driver',
            u'avg_rating_of_driver',u'avg_surge',
            u'surge_pct',u'trips_in_first_30_days',
            u'uber_black_user',u'weekday_pct',
            u'Astapor',u'King\'s Landing',u'Winterfell', # cateogorical feature 'city'
            u'iPhone',u'Android',u'other', # categorical feature 'phone'
            u'user_history' # categorical feature 'signup_date'
           ]

numeric_features = [u'avg_dist',u'avg_rating_by_driver',
                    u'avg_rating_of_driver',u'avg_surge',
                    u'surge_pct',u'trips_in_first_30_days',
                    u'weekday_pct',
                    u'user_history'
                   ]
```

* Split 70% of the data for training and 30% for testing

```python
# Z-Scaling
scaler = preprocessing.StandardScaler().fit(data[numeric_features])
data[numeric_features] = scaler.transform(data[numeric_features])

train_x,  test_x, train_y, test_y = train_test_split(data[features], data['Active'], test_size=0.3)
train_x.shape, test_x.shape
```

* Feature importance

```python
def variable_importance(_model_, x_, y_, fnames, model_name='Random Forest', col='r', TOP_N=50):

    _model_.fit(x_, y_)

    if model_name == 'Random Forest':
        importances = _model_.feature_importances_
        std = np.std([tree.feature_importances_ for tree in _model_.estimators_],
                     axis=0)
    elif model_name == 'Linear Regression':
        importances = np.abs(_model_.coef_)
        std = np.array([0.0] * importances.shape[0])
    else:
        raise ValueError('The model is beyond the current support model list for variable importance!')
    
    indices_ = np.argsort(importances)[::-1]

    # Print the feature ranking
    print("Feature ranking:")

    for f in range(TOP_N):
        print("%d. feature %d -- %s (%f)" % (f + 1, indices_[f], fnames[indices_[f]], 
                                             importances[indices_[f]]))

    # Plot the feature importances of the forest
    fig, ax = plt.subplots(figsize=(15, 10))

    plt.title("{} Feature Importances".format(model_name), fontsize=20)
    plt.bar(range(TOP_N), importances[indices_[:TOP_N]],
            color=col, yerr=std[indices_[:TOP_N]], align="center")

    plt.xticks(range(TOP_N), [name.decode('utf8') 
                              for name in list(np.array(fnames)[indices_])], fontsize=12)
    ax.set_xticklabels([name.decode('utf8') 
                        for name in list(np.array(fnames)[indices_])], rotation=90)
    plt.yticks(fontsize=12)
    ax.set_ylabel('Feature Importance', fontsize=15)

    plt.xlim([-1, TOP_N])
    plt.show()
    
    return indices_
```

```python
rf = RandomForestClassifier(n_estimators=400, n_jobs=4, min_samples_split=10)
variable_indices = variable_importance(rf, train_x, train_y, fnames=features, TOP_N=12)
```

* Model Training and Performance Evaluation

```python
logistic = LogisticRegression(tol=1e-8, C=1e4)
logistic.fit(train_x[variable_indices[:12]], train_y)

prediction = logistic.predict_proba(test_x[variable_indices[:12]])
```

```python
from sklearn.metrics import roc_curve, auc

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)


fpr, tpr, thres = roc_curve(test_y, prediction[:,1])
mean_tpr += np.interp(mean_fpr, fpr, tpr)
mean_tpr[0] = 0.0
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, lw=1, label='ROC (auc = %0.2f)' % (roc_auc))

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic of Rider Retention Classifier', y=1.02)
plt.legend(loc="lower right")
plt.show()
```

```python
rf = RandomForestClassifier(n_estimators=400, n_jobs=4, min_samples_split=20)
rf.fit(train_x[variable_indices[:12]], train_y)

prediction = rf.predict_proba(test_x[variable_indices[:12]])
```
